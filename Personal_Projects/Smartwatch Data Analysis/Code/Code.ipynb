```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.graph_objects as go

data = pd.read_csv("/Users/lekhmac/Documents/GitHub/Coding/Personal_Projects/Smartwatch Data Analysis/Datasets/dailyActivity_merged.csv")
print(data.head())
```

```python
#The column containing the date of the record is an object. We may need to use dates in our analysis, so letâ€™s convert this column into a datetime column:
```

```python
# Changing datatype of ActivityDate
data["ActivityDate"] = pd.to_datetime(data["ActivityDate"], 
                                      format="%m/%d/%Y")
print(data.info())
```

```python
#If we take a look on all the columns, we can see the info about very active, fairly active, light active, and sedentary minutes in the dataset. Let's combine all these columns as total minutes before going ahead
```

```python
data["TotalMinutes"] = data["VeryActiveMinutes"] + data["FairlyActiveMinutes"] + data["LightlyActiveMinutes"] + data["SedentaryMinutes"]
print(data["TotalMinutes"].sample(5))
```

```python
#The dataset has a "calories" column, it contains the data about the number of calories burned in a day. Lets have a look at the realtionship between calories burned and the total steps walked in a day
```

```python
pip install statsmodels
```

```python
figure = px.scatter(data_frame = data, x="Calories",
                    y="TotalSteps", size="VeryActiveMinutes", 
                    trendline="ols", 
                    title="Relationship between Calories & Total Steps")
figure.show()
plt.savefig("figure.png")
```

```python
#In our Plotly Express scatter plot code, the `trendline` parameter is used to add a trendline to the scatter plot. A trendline is a line that represents the general trend or pattern in the data. In our case, you've set `trendline="ols"`, which stands for Ordinary Least Squares. 
#Ordinary Least Squares (OLS) is a method for finding the best-fitting linear regression line through a set of data points. The regression line represents the relationship between the independent variable (x-axis, "Calories" in our case) and the dependent variable (y-axis, "TotalSteps" in our case) in a way that minimizes the sum of the squared differences between the observed and predicted values.
#So, by setting `trendline="ols"`, you're adding a linear regression trendline to our scatter plot, which helps visualize the overall trend or correlation between the "Calories" and "TotalSteps" variables. The trendline shows the general direction and strength of the relationship between these two variables.
```

```python
#Now we have linear relationship between the total number of steps and the number of calories burned in a day. Now lets take a look at the average number of active. minutes burned in a day
```

```python
label = ["Very Active Minutes", "Fairly Active Minutes", 
         "Lightly Active Minutes", "Inactive Minutes"]
counts = data[["VeryActiveMinutes", "FairlyActiveMinutes", 
               "LightlyActiveMinutes", "SedentaryMinutes"]].mean()
colors = ['yellow','green', "purple", "red"]

fig = go.Figure(data=[go.Pie(labels=label, values=counts)])
fig.update_layout(title_text='Total Active Minutes')
fig.update_traces(hoverinfo='label+percent', textinfo='value', textfont_size=30,
                  marker=dict(colors=colors, line=dict(color='black', width=3)))
fig.show()
plt.savefig("figure.png")
```

```python
#STATS:
#   1) 81.3% of Total inactive minutes in a day
#   2) 15.8% of Light active minutes in a day
#   3) On an average, only 21 minutes(1.74%) were very active
#   4) And 1.11%(13 minutes) of fairly active minutes in a day
```

```python
#Recollecting above, we have transformed the data type of the "ActivityDate" column to datetime column above.
#Lets use it to find the weekdays of the records and add a new column to this dataset as "Day"
```

```python
data["Day"] = data["ActivityDate"].dt.day_name()
print(data["Day"].head())
```

```python
#Now letâ€™s have a look at the very active, fairly active, and lightly active minutes on each day of the week
```

```python
fig = go.Figure()
fig.add_trace(go.Bar(
    x=data["Day"],
    y=data["VeryActiveMinutes"],
    name='Very Active',
    marker_color='red'
))
fig.add_trace(go.Bar(
    x=data["Day"],
    y=data["FairlyActiveMinutes"],
    name='Fairly Active',
    marker_color='blue'
))
fig.add_trace(go.Bar(
    x=data["Day"],
    y=data["LightlyActiveMinutes"],
    name='Lightly Active',
    marker_color='black'
))
fig.update_layout(barmode='group', xaxis_tickangle=-45)
fig.show()
plt.savefig("figure.png")
```

```python
#Now lets look at the number of inactive minutes on each day of the week
```

```python
day = data["Day"].value_counts()
label = day.index
counts = data["SedentaryMinutes"]
colors = ['gold','lightgreen', "pink", "blue", "skyblue", "cyan", "orange"]

fig = go.Figure(data=[go.Pie(labels=label, values=counts)])
fig.update_layout(title_text='Inactive Minutes Daily')
fig.update_traces(hoverinfo='label+percent', textinfo='value', textfont_size=30,
                  marker=dict(colors=colors, line=dict(color='black', width=3)))
fig.show()
plt.savefig("figure.png")
```

```python
#As per the above stats we can decide that Thrusday is the most inactive day. Now lets a look at the number of calories burned in each day of the week
```

```python
calories = data["Day"].value_counts()
label = calories.index
counts = data["Calories"]
colors = ['gold','lightgreen', "pink", "blue", "skyblue", "cyan", "orange"]

fig = go.Figure(data=[go.Pie(labels=label, values=counts)])
fig.update_layout(title_text='Calories Burned Daily')
fig.update_traces(hoverinfo='label+percent', textinfo='value', textfont_size=30,
                  marker=dict(colors=colors, line=dict(color='black', width=3)))
fig.show()
plt.savefig("figure.png")
```

```python
#We can conclude that Tuesday is the most active days for all the individuals in the dataset as the highest number of calories burned is on Tuesdays
```
